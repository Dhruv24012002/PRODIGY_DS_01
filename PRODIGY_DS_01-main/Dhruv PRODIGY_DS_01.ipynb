```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python

url=r"E:\API_SP.POP.TOTL_DS2_en_csv_v2_213150\API_SP.POP.TOTL_DS2_en_csv_v2_213150.csv"
df = pd.read_csv(url, skiprows=4)
```

```python
df.head()
```

```python
df = df[['Country Name', 'Country Code', '1960', '1965', '1970', '1975', '1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015', '2020']]
```

```python
selected_year = '1960'
```

```python
df_filtered = df.dropna(subset=[selected_year])
```

```python
df_sorted = df_filtered.sort_values(by=selected_year, ascending=False)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(df_sorted['Country Name'][:10], df_sorted[selected_year][:10], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.title(f'Top 10 Countries by Total Population in {selected_year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
